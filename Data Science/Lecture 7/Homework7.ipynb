```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from skimage.io import imread
from collections import Counter
import networkx as nx
```

```python
sofia = {"lat": 42.6977, "lng": 23.3219}

m = Basemap(projection = "merc", llcrnrlat = 41, llcrnrlon = 22, urcrnrlat = 44.5, urcrnrlon = 29, resolution = "l")
m.drawcoastlines()
m.drawcountries()
x, y = m(sofia["lng"], sofia["lat"])
m.plot(x, y, "o")
plt.show()
```
